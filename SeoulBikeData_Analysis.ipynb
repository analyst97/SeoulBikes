# Transform Data

```python
df_data['DateTime'] = df_data['Date'].astype(str) + ' ' + df_data['Hour'].astype(str)
df_data.head()
```

# Transform Data

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

# encode all features for the machine learning model which are not numeric already
non_numeric = ['Seasons', 'Holiday', 'Functioning Day']

for column in non_numeric:
    df_data[column] = encoder.fit_transform(df_data[column])
    
df_data.head()
```

# Regression Model

```python
holiday_date = df_data.groupby('Date').agg({'Holiday': 'mean'})
holiday_date['Holiday'].unique()
```

```python
functioning_date = df_data.groupby('Date').agg({'Functioning Day': 'mean'})
functioning_date['Functioning Day'].unique()
```

```python
# exclude season as it can be explained by the other factors
```

```python
X_daily = pd.concat([temperature_date, humidity_date, wind_date, visibility_date, solar_date, rainfall_date, snowfall_date, holiday_date, functioning_date], axis=1)
X_daily
```

```python
X_daily_with_bike_rent = pd.concat([X_daily, bike_count_date], axis=1)
X_daily_with_bike_rent.corr()
```

```python
y_daily = bike_count_date['Rented Bike Count']
y_daily
```

```python
from sklearn.model_selection import train_test_split

# split data into training and test set
x_train, x_test, y_train, y_test = train_test_split(X_daily, y_daily, test_size=0.2, shuffle=False, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
y_pred = pd.Series(y_pred)
y_pred
```

```python
y_pred.index = x_test.index
y_pred
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18,5))

y_daily.plot(label='y_test')
y_pred.plot(label='y_pred', color='red')
```

# Base Prediction

```python
y_normal = reg.predict(X_daily)
y_normal = pd.Series(y_normal)
y_normal.index = X_daily.index
```

```python
plt.figure(figsize=(18,5))

y_daily.plot(label='y_test')
y_normal.plot(label='y_pred', color='red')
```

# Sensitivity Analysis

```python
# Scenarios:
# - 70% Probability of +2°C of average temperature ceteris paribus.
# - 30% Probability for an uptake of 3 percentage points in the average humidity ceteris paribus.
```

# Scenario 1

```python
X_future = X_daily
#X_future.index = X_daily.index + pd.offsets.DateOffset(years=1)
X_future['Temperature(°C)'] = X_future['Temperature(°C)'] + 2
X_future
```

```python
y_future = reg.predict(X_future)
y_future = pd.Series(y_future)
y_future.index = X_future.index
```

```python
plt.figure(figsize=(18,5))

y_daily.plot(label='y_test')
y_future.plot(label='y_pred', color='red')
```

# Scenario 2

```python
X_future2 = X_daily
#X_future2.index = X_daily.index + pd.offsets.DateOffset(years=1)
X_future2['Humidity(%)'] = X_future2['Humidity(%)'] + 3
X_future2
```

```python
y_future2 = reg.predict(X_future2)
y_future2 = pd.Series(y_future2)
y_future2.index = X_future2.index
```

```python
plt.figure(figsize=(18,5))

y_daily.plot(label='y_test')
y_future2.plot(label='y_pred', color='red')
```
